```python
from os import replace
import pandas as pd

# URL containing the table
url = "https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"

# Read tables from the URL
tables = pd.read_html(url)
print(tables)
df=tables[3]
df
print(df.head(10))  # Display the first few rows
print(df.columns)  # Display the column names
print("Columns:", df.columns.tolist())
#cleaning the columns names
df.columns = [(col[0].strip(), col[1].strip()) for col in df.columns]
# now  stripping the columns we actually needs
print(df.columns)
print(df.columns[2:5])
#making  df for  needed values
df = df[[df.columns[0]] + list(df.columns[2:4])]
print(df)
 #Rename columns as necessary
df.columns = ["Country", "GDP (Million USD)","Year"]
df.columns[2]
print(df.head(2))
# retaining the top  10 countriess with high gdp
df = df.iloc[1:11, :]
print(df)
## Check the current column names
print(df.columns)
#converting gdp to  float divede by 1000 and then in integer
df['GDP (Million USD)'] = df['GDP (Million USD)'].astype(float)
df['GDP (Million USD)'] = df['GDP (Million USD)'].astype(int)
#converting USD million into billion
df['GDP (Million USD)'] = df['GDP (Million USD)']/1000
df.columns = ["Country", "GDP (Billion USD)","Year"]
print(df)
```
