```python
from selenium import webdriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import csv
import time
```

```python
driver = webdriver.Chrome()
driver.get('https://www.glassdoor.co.in/Job/hyder%C4%81b%C4%81d-data-analyst-jobs-SRCH_IL.0,9_IC2865319_KO10,22.htm')
```

```python
data = []
while True:
    jobs = driver.find_elements(By.CLASS_NAME,'JobCard_jobCardContainer__arQlW')
    for j in jobs:
        try:
            company = j.find_element(By.CLASS_NAME,'EmployerProfile_compactEmployerName__9MGcV').text
        except:
            company = 'N/A'

        try:
            role = j.find_element(By.CLASS_NAME,'JobCard_jobTitle__GLyJ1').text
        except:
            role ='N/A'

        try:
            location = j.find_element(By.CLASS_NAME,'JobCard_location__Ds1fM').text
        except:
            location ='N/A'

        try:
            salary = j.find_element(By.CLASS_NAME,'JobCard_salaryEstimate__QpbTW').text
        except:
            salary = 'N/A'

        try:
            update = j.find_element(By.CLASS_NAME,'JobCard_listingAge__jJsuc').text
        except:
            update = 'N/A'

        data.append({
            'Company Name': company,
            'Role': role,
            'Location': location,
            'Salary' : salary,
            'Last Update':update
        })     
    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "button[data-test='load-more']"))
        )
        next_button.click()
    except:
        break
```
